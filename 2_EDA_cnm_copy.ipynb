##### Manual chi-square calculations

```python
myCrosstable = pd.crosstab(cnm_VarsOnly['Sex'], cnm_VarsOnly['SSSF'])
myCrosstable
```

```python
colTotals = myCrosstable.sum()
rowTotals = myCrosstable.sum(axis=1)

colTotals, rowTotals
```

```python
nCols = len(colTotals)
nRows = len(rowTotals)

nCols, nRows
```

```python
N = rowTotals.sum()
N
```

```python
for i in range(nRows):
    for j in range(nCols):
        E = rowTotals[i] * colTotals[j] / N
        print(E)
```

```python
chiVal = 0
for i in range(nRows):
    for j in range(nCols):
        E = rowTotals[i] * colTotals[j] / N
        O = myCrosstable.iloc[i,j]
        chiVal = chiVal + (O - E)**2 / E
chiVal
```

```python
df = (nRows - 1) * (nCols - 1)
df
```

```python
# !pip install scipy
from scipy.stats import chi2
```

```python
chi2.cdf(chiVal, df)
```

```python
1 - chi2.cdf(chiVal, df)
```

```python
chi2.sf(chiVal, df)
```
